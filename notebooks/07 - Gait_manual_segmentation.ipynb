### Manual segmentation of gait data

Here we will demonstrate how to add manual segmentation labels from videos or any other source on top of pdkit's GaitTimeSeries.

```python
# this is required for plotting in jupyter notebook
%matplotlib inline
```

```python
from pdkit.gait_time_series import GaitTimeSeries
from pdkit.gait_processor import GaitProcessor
import numpy as np
import pandas as pd
```

Let's load the data, from a standard cloudUPDRS format file

```python
ts = GaitTimeSeries().load_data('../tests/data/cloudupdrs_gait.csv')
```

```python
ts.index
```

```python
# instantiate the GaitProcessor
gp = GaitProcessor()
```

Currently we add segmentation data using a python dictionary.

- The keys represent the labels and the list of tuples represent the start and end of each section of that particular label.
- The list of tuples can be of two types, one that uses integers for actual index locations on the timeseries, and another using timestamps.
- If we use integers they will be automatically transformed into timestamps to match the time of the gait data.

```python
manual_segmentation_raw = {'walk': [(100, 450), (558, 962)],
                           'turn': [(68, 98), (455, 551)],
                           'stand': [(0, 65), (982, 1000)]}
```

```python
# utility function to view the segmentation
gp.plot_segmentation_dictionary(ts, manual_segmentation_raw)
```

```python
# if we are happy with the segmentation we can add it to the dataframe
segmented_df = gp.add_manual_segmentation_to_data_frame(ts, manual_segmentation_raw)
```

```python
# as you can see we have a new column with segmentation data!
segmented_df.head()
```

```python
# utility function to view the segmented dataframe
gp.plot_segmentation_data_frame(segmented_df)
```

We can now do the exact same thing using datetime indices instead of numeric indexes.

```python
manual_segmentation_raw_timed = {'walk': [(np.datetime64('1970-01-01T00:00:02.124756000'), np.datetime64('1970-01-01T00:00:09.217316000')),
                                        (np.datetime64('1970-01-01T00:00:11.414276000'), np.datetime64('1970-01-01T00:00:19.439697000'))],
                               'turn': [(np.datetime64('1970-01-01T00:00:01.462250000'), np.datetime64('1970-01-01T00:00:02.082459000')),
                                        (np.datetime64('1970-01-01T00:00:09.320923000'), np.datetime64('1970-01-01T00:00:11.284546000'))],
                               'stand': [(np.datetime64('1970-01-01T00:00:00.000000000'), np.datetime64('1970-01-01T00:00:01.402985000')),
                                         (np.datetime64('1970-01-01T00:00:20.035370000'), np.datetime64('1970-01-01T00:00:20.409302000'))]}
```

```python
gp.plot_segmentation_dictionary(ts, manual_segmentation_raw_timed)
```

```python
segmented_df_timed = gp.add_manual_segmentation_to_data_frame(ts, manual_segmentation_raw_timed)
segmented_df_timed.head()
```

```python
gp.plot_segmentation_data_frame(segmented_df_timed)
```

Another interesting aspect of using numerical segmentation values instead of datetime is that they are not portable to resampled data.

Let's resample the dataframe at 100hz and see how that works

```python
resampled = gp.resample_signal(ts)
filtered = gp.filter_data_frame(resampled)
```

```python
segmented_resampled_df = gp.add_manual_segmentation_to_data_frame(resampled, manual_segmentation_raw)
```

```python
gp.plot_segmentation_data_frame(segmented_resampled_df)
```

Instead if we use the datetime segmentation we will get the real segmentation.

```python
segmented_timed_resampled_df = gp.add_manual_segmentation_to_data_frame(resampled, manual_segmentation_raw_timed)
```

```python
gp.plot_segmentation_data_frame(segmented_timed_resampled_df)
```

If you wanted to use numerical segmentation you have to provide the real segmentation values for the resampled data.

```python
manual_segmentation_filtered = {'walk': [(170, 997), (1155,1992)],
                               'turn': [(1004, 1152)],
                               'stand': [(0, 165), (2002, 2320)]}
```

```python
gp.plot_segmentation_dictionary(filtered, manual_segmentation_filtered)
```

```python
segmented_filtered = gp.add_manual_segmentation_to_data_frame(filtered, manual_segmentation_filtered)
```

```python
gp.plot_segmentation_data_frame(segmented_filtered)
```

### Lastly let's save the segmented data

For this we will use the builtin pandas saving functionality for DataFrames

```python
segmented_timed_resampled_df.to_csv('/tmp/filtered_segmented_data.csv')
```

And now let's load it back

```python
seg_ts = GaitTimeSeries.load_data('/tmp/filtered_segmented_data.csv', format_file='segmented')
```

```python
gp.plot_segmentation_data_frame(seg_ts)
```
