```python
from pdkit import GaitTimeSeries, GaitProcessor
from pdkit.utils import plot_segmentation
```

```python
%matplotlib inline
import numpy as np
```

```python
import matplotlib.pylab as plt
```

```python
data = '/src/data/pdkit/physics_phone_data/wtwtwlpfd.csv'
```

```python
ts = GaitTimeSeries.load_data(data, format_file='physics')
```

```python
ts.mag_sum_acc.plot()
```

```python
gp = GaitProcessor(sampling_frequency=50.0)
```

```python
rts = gp.resample_signal(ts)
rts.mag_sum_acc.plot()
```

```python
frts = gp.filter_data_frame(rts)
```

```python
frts.mag_sum_acc.plot()
```

```python
gp.axis = 'mag_sum_acc'
```

```python
peak, promi, segments = gp.bellman_segmentation(frts, 8)
```

```python
plt.plot(frts.mag_sum_acc[peak])
```

```python
plot_segmentation(np.sqrt(frts.mag_sum_acc.values), peak, segments)
```

```python
plt.plot(np.sqrt(frts.mag_sum_acc.values))
```

```python
dd = gp.filter_signal(rts)
```

```python
gp.axis = 'filtered_signal'
```

```python
pk, prm, idx = gp.bellman_segmentation(dd, 10)
```

```python
plot_segmentation(dd.filtered_signal.values, pk, idx)
```

```python
from sklearn.cluster import Birch
```

```python
cluster_fn = Birch(n_clusters=2)
```

```python
peak, promi, segments = gp.sklearn_segmentation(dd, cluster_fn=cluster_fn)
```

```python
plot_segmentation(dd.filtered_signal.values, peak, segments)
```

```python
from pdkit.utils import BellmanKSegment, plot_segmentation
```

```python
clstr = BellmanKSegment(frts.mag_sum_acc, 8)
```

```python
def Display(data, ix):
    plt.plot(data);
    for segment in np.unique(ix):
        plt.plot(np.where(ix == segment)[0],data[np.where(ix == segment)[0]],'o')
    plt.show()
```

```python
Display(frts.mag_sum_acc.values, clstr)
```
