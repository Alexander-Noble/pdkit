# IC + FC

following: **An enhanced estimate of initial contact and final contact instants of time using lower trunk inertial sensor data** paper

`doi:10.1016/j.gaitpost.2012.02.019`

```python
from pdkit.gait_time_series import GaitTimeSeries
from pdkit.gait_processor import GaitProcessor
from pdkit.utils import peakdet
```

```python
%matplotlib inline
import pywt
import matplotlib.pylab as plt
import numpy as np
```

```python
ts = GaitTimeSeries.load_data('../tests/data/cloudupdrs_gait.csv')
gp = GaitProcessor()
```

### Get the Initial Contact (IC) locations

```python
coef, freqs = pywt.cwt(ts.x.values, scales=range(1, 16), wavelet='gaus1')
```

```python
ic = coef[14]
plt.plot(ic)
```

```python
maxp, minp = peakdet(ic, delta=3)
```

```python
plt.plot(ic)
plt.scatter(minp[:, 0], minp[:, 1], color='red')#
plt.scatter(maxp[:, 0], maxp[:, 1], color='green')
```

### Get the Final Contact (FC) locations

```python
coef, freqs = pywt.cwt(ic, scales=range(1, 16), wavelet='gaus1')
```

```python
ic = coef[14]
plt.plot(ic)
```

```python
maxp, minp = peakdet(ic, delta=3)
```

```python
plt.plot(ic)
plt.scatter(minp[:, 0], minp[:, 1], color='red')
plt.scatter(maxp[:, 0], maxp[:, 1], color='green')
```
